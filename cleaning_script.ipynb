```python
import pandas as pd
import numpy as np

# 1. شارجي الداتا (ما تنساش point-virgule)
df = pd.read_csv('student-mat.csv', sep=';')

# 2. شوف واش كاينين قيم ناقصة (Missing Values)
print("عدد القيم الناقصة قبل التنقية:")
print(df.isnull().sum())

# 3. التنقية (Cleaning)
# هي نورمالمون الداتا ديال UCI كتكون نقية، ولكن احتياطاً
# إلا لقينا شي سطر فيه Null، غانمسحوه
df = df.dropna()

# 4. سجل الملف الجديد (باش يخدم عليه صاحبك)
df.to_csv('temp_data_cleaned.csv', index=False)

print("✅ صافي ساليتي! الملف 'temp_data_cleaned.csv' واجد.")
```
